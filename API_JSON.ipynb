**Importing and saving the movies dataset (best practice)**

```python
import random
randomlist = []
for i in range(0,20):
    n = random.randint(0,300000)
    randomlist.append(n)
print(randomlist)
```

```python
movie_id = randomlist
```

```python
basic_url = 'https://api.themoviedb.org/3/movie/{}?{}'
```

```python
json_list = []
for movie in movie_id:
    url = basic_url.format(movie, api_key)
    r = requests.get(url)
    if r.status_code != 200:
        continue
    else:
        data = r.json()
        json_list.append(data)

df = pd.DataFrame(json_list)   
```

```python
#Showing the results when there is no data available
requests.get(basic_url.format(0, api_key)).status_code
```

```python
df
```

```python
df = df.loc[:, ['title', 'id', 'revenue', 'genres', 'belongs_to_collection', 'runtime']].sort_values(ascending=False, by='revenue')
df
```

```python
df.to_json('movies.json', orient='records')
```
